```python
import numpy as np
import xlrd
#打開文件
data = xlrd.open_workbook('training.xlsx')
#打開Excel工作表1
table = data.sheets()[0]
nrows = table.nrows #行数
ncols = table.ncols #列数
print(table)
print(nrows)
print(ncols)
 
start=0  #开始的行
end=183  #结束的行
 
rows=end-start
 
list_values=[] #儲存所有的值
for x in range(start,nrows):
    values=[]  #儲存單列的值
    # 讀入第x列的所有值
    row =table.row_values(x)

    for i in range(0,9):
        #將單列第i個值(row[i]) 放到value[]後面
        values.append(row[i])
        #將第x列 value 放到list_values[]後面
    list_values.append(values)
       
    
#傳入list到 numpy.array() 創建陣列  會排版
datamatrix=np.array(list_values)
print(datamatrix)

```

```python
list_values=[] #儲存所有的值
for x in range(0,9):
    values=[]  #儲存單行的值
    # 讀入第x行的所有值
    col =table.col_values(x)
    for i in range(0,183):
        #將單列第i個值(col[i]) 放到value[]後面
        values.append(col[i])
        #將第x列 value 放到list_values[]後面
    list_values.append(values)
    
    
#傳入list到 numpy.array() 創建陣列  會排版
datamatrix=np.array(list_values)
print(datamatrix)
```

```python
#印高關懷那行
grade=[]
for i in range(1,183):
    grade.append(datamatrix[8][i])
grade = list(map(float,grade))
print(grade)
```

```python
#以三科不及格當作高關懷數量
for x in range(1,nrows):
    row =table.row_values(x)
    fail=[]
    try:
        
        row = list(map(float,row))
        
    except ValueError:
        pass
    
for x in range(1,183):
    not_pass=0 
    row =table.row_values(x)
    
    
    for i in range(1,6):
        if row[i]==''or row[i]==' ':
            pass
        elif row[i]<60:
            not_pass=not_pass+1
          
    if not_pass>=3:
        fail.append(1)
    else:
        fail.append(0)
    
print(fail)
```

```python
#以三科不及格下去對比正確率
count =0
for i in range(0,181):
    if fail[i] == grade[i]:
        count+=1
print(count)
percent = count/182*100
print(str(percent)+'%')
```
